### Coal Production Rankings

```python
import pandas as pd

# Load the dataset from the Excel file
df = pd.read_excel("Datasets/CoalProduction.xlsx")

# Calculate average coal production for each country
average_coal_production = df.groupby("Entity")["Coal Production(TWh)"].mean()

# Rank the countries based on average coal production
ranked_countries = average_coal_production.rank(ascending=False)

# Create a dictionary to store country ranks
coalProductionRank = dict(zip(ranked_countries.index, ranked_countries))

print(coalProductionRank)
```

### Ease of Doing Business Rankings

```python
import pandas as pd

# Load the dataset from the Excel file
df = pd.read_excel("Datasets/EaseOfDoingBusiness.xlsx")

# Calculate average score for each country
df["Average Score"] = df[["2019", "2020"]].mean(axis=1)

# Sort the dataframe based on the average score in descending order
df_sorted = df.sort_values(by="Average Score", ascending=False)

# Create a dictionary to store country rankings
EaseOfDoingBusinessRank = {}

# Assign rankings to countries based on the average score
rank = 1
for _, row in df_sorted.iterrows():
    EaseOfDoingBusinessRank[row["Country"]] = rank
    rank += 1

# Print the country rankings
print(EaseOfDoingBusinessRank)
```

### Working Age Population Ranking

## Average Ranking of Countries

```python
# Create a list of all the dictionaries
rankings = [coalProductionRank, CrudeOilProductionRank, EaseOfDoingBusinessRank, EmploymentRateRank, GDPperCapitaRank, GlobalInnovationIndexRank, GlobalKnowledgeIndexRank, HealthExpenditureRank, InfrastructureScoreRank, PoliticalStabilityRank, QualityOfLifeRank, WorkingAgePopulationRank]

# Create an empty dictionary to store the average ranks
average_ranks = {}

# Create a set to store all the countries
countries = set()

for rank_dict in rankings:
    # Add all the countries in the current dictionary to the countries set
    countries.update(rank_dict.keys())

for country in countries:
    # Initialize variables to store the sum of ranks and the number of valid ranks for the current country
    rank_sum = 0
    valid_ranks = 0
    
    for rank_dict in rankings:
        # Retrieve the rank of the current country in the current dictionary using the get method and specify a default value of None
        rank = rank_dict.get(country, None)
        
        # Check if the rank is not None
        if rank is not None:
            # Add the rank to the rank_sum variable and increment the valid_ranks variable
            rank_sum += rank
            valid_ranks += 1
    
    # Check if valid_ranks is greater than 0
    if valid_ranks > 0:
        # Calculate the average rank for the current country and round it to 3 decimal places using round function and store it in the average_ranks dictionary
        average_ranks[country] = round(rank_sum / valid_ranks, 3)

# Print the average_ranks dictionary
print(average_ranks)
```

## Probability Of Becoming World Economic Center (Out of 100)

```python
adjusted_ranks = {country: (20 - rank) for country, rank in average_ranks.items()}
total = sum(adjusted_ranks.values())
shares = {country: (rank / total) * 100 for country, rank in adjusted_ranks.items()}
print(shares)
```

### Probability Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a larger figure
fig1, ax1 = plt.subplots(figsize=(9, 9))

# Create the pie chart
wedges, texts, autotexts = ax1.pie(shares.values(), labels=shares.keys(), autopct='%1.1f%%', startangle=90)

# Adjust the autotext rotation
for autotext in autotexts:
    autotext.set_rotation_mode('anchor')
    autotext.set_rotation(-autotext.get_position()[0] * 1 / np.pi)

# Set the aspect ratio to equal to make the pie circular
ax1.axis('equal')

# Show the pie chart
plt.show()
```
